# World Economic Indicators: Analyzing Global Development Trends

## Overview

- This project analyzes global economic and development trends using datasets from the World Bank and United Nations (UN) Human Development Index (HDI).

- The dataset covers economic indicators from 1960-2018, and HDI data from 1990-2021.

- The analysis explores GDP growth, HDI trends, life expectancy correlations, and income disparities.

## Data Loading & Preprocessing

```python
import pandas as pd
import numpy as np

# Loading the dataset
world_bank = pd.read_excel("/Users/venkat/Downloads/World Bank Economic Report/World+Economic+Indicators/WorldBank.xlsx").rename({"Country Name":"Country"},axis=1)

world_bank.head()
```

```python
world_bank["Population (M)"] = (world_bank["GDP (USD)"]/world_bank["GDP per capita (USD)"]/1000000)
```

```python
world_bank.head()
```

```python
world_bank.info()
```

```python
world_bank.describe()
```

To expand the scope of already existing World Bank data, Human Development Index(HDI) data has also been provided by the United Nations.
This supplementary dataset contains additional indicators used by the UN to track development, environmental impact, and inequality for each country from 1990-2021.

```python
hdi = pd.read_csv("/Users/venkat/Downloads/World Bank Economic Report/World+Economic+Indicators/hdi.csv")

hdi.head()
```

```python
# Filtering the data and joining both hdi and world bank data on "Country Code" column.
# The reason for picking hdi_2014 is World Bank data has lot of information on the statistics we are interested as a part of this project for 2014. For other years like 2015 and 2016 , there is a lot of missing data on electrcity consumption which might make inconsistent visualizations down the line. So the two data tables with respect to this project have been joined on hdi_2014. 
wb_hdi_2014 = world_bank.query("Year==2014").merge(hdi[["iso3","hdi_2014"]] , how="left", left_on="Country Code",right_on="iso3")

wb_hdi_2014.head()
```

```python
wb_hdi_2014.describe()
```

## Data Transformations

```python
# Creating a gdp_pivot table with years as rows and regions as columns , sum of GDP as values.

gdp_pivot = world_bank.pivot_table(index="Year", columns="Region",values = "GDP (USD)",aggfunc="sum")

gdp_pivot.head()
```

```python
# Creating pop_pivot table with years as rows , regions as columns and sum of population as values.

pop_pivot = world_bank.pivot_table(index="Year", columns="Region", values = "Population (M)",aggfunc="sum")

pop_pivot.head()
```

```python
# Creating a "wb_hdi_by_region" table by calculating the average HDI for each region and sorting from highest average HDI to lowest.

wb_hdi_by_region = wb_hdi_2014.groupby("Region").agg({"hdi_2014": "mean"}).sort_values("hdi_2014", ascending=False)

wb_hdi_by_region.head()
```

## Data Visualization

The objective here is to create charts to visualize trends, relationships and comparisons across regions.

```python
# Here we are creating a stacked area chart showing the growth of GDP over time with each stack representing a geographic region.

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

ax.stackplot(
    gdp_pivot.index,
    [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
    labels =gdp_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.legend(loc="upper left")
ax.set_title("GDP has grown exponentially over time")
ax.set_ylabel("GDP(trillions)")
```

```python
gdp_pivot.iloc[-1].sort_values(ascending=False).index
```

```python
# Next creating a stacked area chart showing the growth of population over time, with each stack representing a geographic region

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots()

ax.stackplot(
    pop_pivot.index,
    [pop_pivot[region]/1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels =gdp_pivot.iloc[-1].sort_values(ascending=False).index
)

ax.set_title("Population has surged from 2 billion to 7.5 billion")
ax.set_ylabel("Population (Billions)")
```

```python
# This is a bubble chart with life expectancy for each industry in the x-axis , GDP per capita in the y-axis and population as the bubble size. 

import matplotlib.ticker as mticker

fig, ax =  plt.subplots()

minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])


sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes = (minsize,maxsize),
    hue="Region",                  # Bubbles have been coloured by region
    ax=ax
).set(title="Life expectancy increases as countries get richer",yscale="log")

ax.yaxis.set_major_formatter(mticker.ScalarFormatter())
```

```python
# Bar chart has been created next to show average HDI by region

wb_hdi_by_region.plot.bar(title="HDI by region",ylabel="Human Development Index(HDI)")
```

```python
wb_hdi_2014.info()
```

```python
# This is a scatter plot with power consumption in the x-axis and GDP per capita in the y-axis. 

sns.scatterplot(data=wb_hdi_2014.query("Country!='Iceland'"),
               x="Electric power consumption (kWh per capita)",
               y="GDP per capita (USD)",
               hue="hdi_2014",     # HDI has been used as color scale
               palette='coolwarm_r').set(title="Electricity Drives Development")
```

## Creating a Report

```python
# Assmeble the visualizations above into a single figure. We will be using Matplotlib's Gridmesh functionality here.

# There will be total 3 rows of visuals. In row 1, we are placing the stacked charts, in row 2 we will place the bubble chart and in row 3, we will place the bar chart and scatterplot side by side at the bottm

import matplotlib.gridspec as gridspec
import matplotlib as mpl

fig = plt.figure(figsize=(8, 12))

gs = gridspec.GridSpec(nrows=12, ncols=6)

stack1_list = ['#66C2A5','#fC8D62', '#8DA0CB', '#E78AC3', '#A6D854', '#FFD92F','#B3B3B3']
stack2_list = ['#66C2A5','#A6D854','#B3B3B3','#fC8D62','#E78AC3','#8DA0CB','#FFD92F']
bar_list = ['#8DA0CB','#fC8D62','#E78AC3','#FFD92F','#66C2A5','#A6D854','#B3B3B3']
bubble_list = ['#A6D854','#fC8D62','#FFD92F','#66C2A5','#B3B3B3','#E78AC3','#8DA0CB']

fig.suptitle("Report: Global Economic Report from 1960-2018",
            fontsize=16,
            fontweight="bold",
            x=.65,
            y=.93)

ax1=fig.add_subplot(gs[0:4,0:3])


ax1.stackplot(gdp_pivot.index,
              [gdp_pivot[region] / 1_000_000_000_000 for region in gdp_pivot.iloc[-1].sort_values(ascending=False).index],
              labels =gdp_pivot.iloc[-1].sort_values(ascending=False).index,
              colors = stack1_list
)


ax1.set_title("GDP has grown exponentially over time")
ax1.set_ylabel("GDP(trillions)")

fig.text(.92,.62,
'''
Economic Expansion in the 20th and early 21st centuries 
has been enormous and has outplaces population growth.
In 1960, Europe and North America produced 75% of GDP on 20% of population.

By 2018 East Asia had surpassed both in economic output
largely due to the rise of China, Japan and South Korea.

In the coming decades, economists expect South Asia 
and Sub-Saharan Africa to lead the world in global growth.'''
        ,fontsize=8)

ax2=fig.add_subplot(gs[0:4, 3:6])

ax2.stackplot(
    pop_pivot.index,
    [pop_pivot[region] / 1000 for region in pop_pivot.iloc[-1].sort_values(ascending=False).index],
    labels =gdp_pivot.iloc[-1].sort_values(ascending=False).index,
    colors = stack2_list
)

ax2.set_title("Population has surged from 2 billion to 7.5 billion")
ax2.set_ylabel("Population (Billions)")
ax2.legend(bbox_to_anchor=(1.75,1),fontsize=8)


ax3=fig.add_subplot(gs[4:8, :])

minsize = min(wb_hdi_2014["Population (M)"])
maxsize = max(wb_hdi_2014["Population (M)"])


sns.scatterplot(
    data=wb_hdi_2014,
    x="Life expectancy at birth (years)",
    y="GDP per capita (USD)",
    size="Population (M)",
    sizes = (minsize,maxsize),
    hue="Region",
    palette = bubble_list,
    ax=ax3
).set(title="Life expectancy increases as countries get richer",yscale="log")

ax3.yaxis.set_major_formatter(mticker.ScalarFormatter())

handles, labels = ax3.get_legend_handles_labels()
entries_to_skip = len(wb_hdi_2014["Region"].unique())+1

for h in handles[1:]:
    sizes = [s / 1.5 for s in h.get_sizes()]
    h.set_sizes(sizes)

ax3.legend(
    handles[entries_to_skip:], 
    labels[entries_to_skip:] ,
    bbox_to_anchor=(.8,.9), 
    loc=2,borderaxespad=0,
    frameon=False,
    ncol=6,
    fontsize=8
)

fig.text(.92,.38,
'''
The wealthy nations of the world enjoy high GDP
per capita as well as long life spans, but make 
up a relatively small share of the global population.


If growth can continue in the developing world, 
humanity will be vastly wealthier, healthier, 
and (hopefully) happier.''',fontsize=8)

ax4=fig.add_subplot(gs[8:12, 0:3])

ax4= wb_hdi_by_region.plot.bar(title="HDI by region", ylabel="Human Development Index(HDI)", color = bar_list)

ax5=fig.add_subplot(gs[8:12, 3:6])

sns.scatterplot(
    data=wb_hdi_2014.query("Country!='Iceland'"),
    x="Electric power consumption (kWh per capita)",
    y="GDP per capita (USD)",
    hue="hdi_2014",
    palette='coolwarm_r',
    ax=ax5).set(title="Electricity Drives Development")

fig.text(.92,.15,
'''
HDI, short for Human Development Index,
attempts to measure the overall standard
of living in countries. Life Expectancy, 
GDP per capita and educational attainment
are the factors that are considered.

Economic growth in the developing world in the 
21st century should help other regions catch up
to North America and Europe. 

One factor not in HDI, but instrumental to a
country's development, is electricity consumption.
Electricity unlocks massive improvements in productivity.
Developing nations should continue to invest in energy
production to ensure growth.
''',
        fontsize=8)

fig.subplots_adjust(wspace=4,hspace=4)
```

```python
correlation = wb_hdi_2014[['GDP per capita (USD)', 'hdi_2014']].corr().loc['GDP per capita (USD)', 'hdi_2014']

print(f"Correlation between GDP per capita and HDI in 2014: {correlation:.4f}")
```

A correlation of 0.6493 between GDP per capita and HDI in 2014 suggests that higher economic prosperity generally leads to better living standards, education, and healthcare, but other factors like income inequality and government policies also influence HDI.
